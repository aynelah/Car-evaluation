## Predict the safety of the car using decision tree

```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('car_evaluation.csv')
df.head()
```

```python
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']


df.columns = col_names

col_names
```

```python
df.head()
```

```python
df.info()
```

```python
y = df.loc[: , 'class'] 
```

```python
x = df.drop(['class'], axis = 1 )
```

```python
x
```

```python
features = pd.get_dummies(x, columns= ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2)
```

```python
from sklearn import tree 
clf = tree.DecisionTreeClassifier(max_depth = 6, min_samples_split=5 , min_samples_leaf=2)
clf.fit(X_train , y_train)
```

```python
y_train_predict = clf.predict(X_train)
y_test_predict = clf.predict(X_test)
from sklearn.metrics import accuracy_score
train_acc= accuracy_score(y_train, y_train_predict)
test_acc= accuracy_score(y_test ,y_test_predict )
```

```python
print('train accuracy = ' , train_acc)
print('test accuracy = ' , test_acc)
```
